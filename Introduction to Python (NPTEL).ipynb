```python
import matplotlib.pyplot as plt
plt.title("The Graph",fontsize = 24,color = 'Red')
plt.xlabel("The Numbers",fontsize = 18,color = 'Blue')
plt.ylabel("The Squares",fontsize = 18,color = 'Green')
plt.grid(True)
x = [i for i in range (11)]
plt.plot(x,x**2,'r-',x,x**3,'g-',x,x**4,'b-')
```

```python
import numpy
import matplotlib.pyplot as plt
plt.title("The Graph",fontsize = 24,color = 'Red')
plt.xlabel("The Numbers",fontsize = 18,color = 'Blue')
plt.ylabel("The Squares",fontsize = 18,color = 'Green')
plt.grid(True)
x = [i for i in range (11)]
x = numpy.array(x)
plt.plot(x,x**2,'r-',x,x**3,'g-',x,x**4,'b-')
```
